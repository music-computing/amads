## <font color='green'> Hamming Distance

This is a measurement simply of the number of positions at which corresponding symbols are different.
It is useful for categorical or binary data.

Let's consider this as applied to a tune represented only as a sequence of pitch names.

```python
# Example with the categorical data of pitch names in a melody
tune_a = np.array(list('CDEFG'))
tune_b = np.array(list('CDEAG'))

# Convert to numeric or use direct comparison
tune_dist = hamming(tune_a, tune_b)
print(f"A: {''.join(tune_a)}")
print(f"B: {''.join(tune_b)}")
print(f"Hamming distance (fraction): {tune_dist}")
print(f"Hamming distance (raw count): {tune_dist * len(tune_a)}")
```

## <font color='green'> P-norm including "Manhattan" and "Euclidean"

Euclidean distance measures the straight-line distance between two points in Euclidean space.

**Formula:** $d(p, q) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2}$

Perhaps the simplest demonstration of this is with a
3-4-5 right-angle triangle
connecting the origin (0, 0)
with a point at either (3, 4) or (4, 3)
and using those axes as "sides".

Where the "sides" are of length 3 and 4, the straight line (hypotenuse) is 5.

```python
point_a = np.array([0, 0])
point_b = np.array([3, 4])

dist_eucl = euclidean(point_a, point_b)
print(f"Point A: {point_a}")
print(f"Point B: {point_b}")
print(f"Euclidean distance: {dist_eucl}")
```

This can be generalised in terms of the "pnorm".

Here is the function for calculating the p-norm distance between two vectors.

$$\left( \sum_{i} |x_i - y_i|^p \right)^{1/p}$$

At the centre of this is:
1. for each term $i$ taking the differences between corresponding entries $x_i - y_i$
2. the value of p denoting different distance measures.

For example, when $p=2$:

$$\left( \sum_{i} |x_i - y_i|^2 \right)^{1/2}$$

This is the Euclidean distance defined above.

```python
print(f"Euclidean distance (AMADS): {norm.euclidean_distance(point_a, point_b)}")
print(f"Euclidean distance (scipy): {euclidean(point_a, point_b)}")
```

Adapting the value of p gives you other distance.

The 'L1' aka 'Manhattan' distance between two points in N dimensional space.
The distance is the sum of the absolute differences along each axis.

This is calculated simply with the above equation and $p=1$.
With $p=1$, the equation simplifies to:

$$\left( \sum_{i} |x_i - y_i| \right)$$



```python
print(f"Manhattan distance (AMADS): {norm.manhattan_distance(point_a, point_b)}")
print(f"Manhattan distance (scipy): {cityblock(point_a, point_b)}")
```

Finally, $p$ can be set to infinity.

This "infinity" norm (or "max" norm) of a vector is simply the largest absolute value among its elements.

$$|x|_\infty = \max(|x_1|, |x_2|, |x_3|, \ldots)$$

Basically, when you raise numbers to high powers, the largest element dominates.

```python
print(f"Max distance (AMADS): {norm.pnorm_distance(point_a, point_b, p=np.inf)}")
print(f"Max distance (scipy): {chebyshev(point_a, point_b)}")
```

## <font color='green'> Summary, Other Distances, Other Names

| Distance | Use Case | Formula |
|----------|----------|--------|
| **Euclidean** | Continuous numeric data (e.g., coordinates) | √(Σ(aᵢ-bᵢ)²) |
| **Hamming** | Categorical or binary strings | Count of mismatches |
| **Manhattan** | Grid-based paths (L1 norm) | Σ|aᵢ-bᵢ| |

- SciPy supports these and many other metrics.
[Click here for a full list.](https://docs.scipy.org/doc/scipy/reference/spatial.distance.html)

```python
p1 = np.array([1, 2, 3])
p2 = np.array([4, 5, 6])

metrics = ['euclidean', 'cityblock', 'cosine', 'chebyshev']

print("Distance comparison:")
for m in metrics:
    d = cdist([p1], [p2], metric=m)[0, 0]
    print(f"  {m}: {d:.4f}")
```

## <font color='green'> Multiple Points at Once Using `cdist`

Finally, scipy's `cdist` allows you to compute distances between two *collections* of points at once.

This `cdist` implementation is efficient (using C under the hood)
and accepts different options for the metric.

Try changing the metric (e.g., `metric='cityblock'`) to see the effect change.

```python
X = np.array([[0, 0], [1, 1], [2, 2]])
Y = np.array([[3, 4], [5, 6]])
```

```python
distances = cdist(X, Y, metric='euclidean')  # << try 'cityblock' and others
print("Distance matrix (rows=X, cols=Y):")
print(distances)
```

Now for an example visualised.

Again, try changing the metric (`metric='cityblock'`) to see the effect change.

```python
A = np.array([[0, 0], [1, 0]])
B = np.array([[0, 1.5], [2, 2]])

# All pairwise distances in one call
dist_matrix = cdist(A, B, metric='cityblock')  # << try other metrics

print("Distance matrix:")
print(dist_matrix)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

# Plot points from set A as blue circles
ax.scatter(A[:, 0], A[:, 1], s=150, c='steelblue', edgecolors='black', zorder=5, label='Set A')
for i, (x, y) in enumerate(A):
    ax.text(x, y - 0.25, f'A{i}', ha='center', fontsize=12, fontweight='bold')

# Now points from set B as orange squares
ax.scatter(B[:, 0], B[:, 1], s=150, c='orange', marker='s', edgecolors='black', zorder=5, label='Set B')
for i, (x, y) in enumerate(B):
    ax.text(x + 0.15, y, f'B{i}', ha='left', fontsize=12, fontweight='bold')

# Connecting lines (pairwise distances)
for i in range(len(A)):
    for j in range(len(B)):
        ax.plot([A[i, 0], B[j, 0]], [A[i, 1], B[j, 1]], 'gray', alpha=0.4, linewidth=1)
        # Distance values along the line (at the midpoint)
        mid_x = (A[i, 0] + B[j, 0]) / 2
        mid_y = (A[i, 1] + B[j, 1]) / 2
        ax.text(mid_x, mid_y, f'{dist_matrix[i, j]:.3f}', fontsize=8, ha='center', color='gray')

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.8, 2.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('cdist: All Pairwise Distances Between Two Point Sets', fontsize=12)
ax.legend(loc='upper left')
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
