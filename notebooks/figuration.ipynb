With three elements there are 6 options.

Here we define those 6 permutation types and assing a colour to each for plotting

```python
PERM_MAP = {
    (1, 2, 3): 'green',   # Ascending
    (1, 3, 2): 'orange',  # Peak type 1
    (2, 1, 3): 'purple',  # Valley type 1
    (2, 3, 1): 'blue',    # Peak type 2
    (3, 1, 2): 'red',     # Valley type 2
    (3, 2, 1): 'grey',    # Descending
}
```

```python
import matplotlib.pyplot as plt

def plot(data):
	"""
	- Get colours for each element in the data
	- Create visualisation: wide, short figure, with tall thin bars per element
	- Label x by bar not index (4 to a bar, move from 0- to 1- indexing)
	"""
	def get_color(perm):
		if perm is None:
			return 'black'
		return PERM_MAP.get(tuple(perm), 'black')

	colors = [get_color(p) for p in data]

	fig, ax = plt.subplots(figsize=(15, 2))

	for i, color in enumerate(colors):
		ax.axvspan(i, i + 1, 0, 1, color=color, lw=0)

	n_samples = len(data)

	tick_locs = range(0, n_samples + 1, 4)
	ax.set_xticks(tick_locs)
	ax.set_xticklabels([str(t // 4 + 1) for t in tick_locs])

	ax.set_xlim(0, n_samples)
	ax.set_ylim(0, 1)

	ax.set_yticks([])
	ax.set_ylabel("")
	ax.set_xlabel("Bar number (1-indexed) with 4 figurations per bar")
	ax.set_title(f"Visual Summary of {n_samples} Permutations (Color Coded by Type)")

	from matplotlib.patches import Patch
	legend_elements = [Patch(facecolor=c, label=str(p)) for p, c in PERM_MAP.items()]
	ax.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(0, 1), ncol=2)

	plt.tight_layout()
	plt.show()
```

```python
plot(chunk_orders)
```

Note how clearly this shows:
- the overall dominance of the 'rising' `(1, 2, 3)` profile
- the score positions in which that changes (development and coda)
