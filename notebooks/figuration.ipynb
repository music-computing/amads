With three elements there are 6 options.

Here we define those 6 permutation types and assing a colour to each for plotting

```python
PERM_MAP = {
	(1, 2, 3): 'green',   # Ascending
	(1, 3, 2): 'orange',  # Peak type 1
	(2, 1, 3): 'purple',  # Valley type 1
	(2, 3, 1): 'blue',    # Peak type 2
	(3, 1, 2): 'red',     # Valley type 2
	(3, 2, 1): 'grey',    # Descending
}
```

```python
import matplotlib.pyplot as plt

def plot(
		data,
		permutation_map: dict = PERM_MAP,
		iterations_per_bar: int = 4
):
	"""
	- Get colours for each element in the data
	- Create visualisation: wide, short figure, with tall thin bars per element
	- Label x by bar not index (4 to a bar, move from 0- to 1- indexing)
	"""
	def get_color(perm, black='black'):
		if perm is None:
			return black
		return permutation_map.get(tuple(perm), 'black')
	colors = [get_color(p) for p in data]

	fig, ax = plt.subplots(figsize=(15, 2))

	for i, color in enumerate(colors):
		ax.axvspan(i, i + 1, 0, 1, color=color, lw=0)

	n_samples = len(data)

	tick_locs = range(0, n_samples + 1, iterations_per_bar)
	ax.set_xticks(tick_locs)
	ax.set_xticklabels([str(t // iterations_per_bar + 1) for t in tick_locs])

	ax.set_xlim(0, n_samples)
	ax.set_ylim(0, 1)

	ax.set_yticks([])
	ax.set_ylabel("")
	ax.set_xlabel(f"Bar number (1-indexed) with {iterations_per_bar} figurations per bar")
	ax.set_title(f"Visual Summary of {n_samples} iterations of the arpeggiation figure (Color Coded by Type)")

	from matplotlib.patches import Patch
	legend_elements = [Patch(facecolor=c, label=str(p)) for p, c in permutation_map.items()]
	ax.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(0, 1), ncol=2)

	plt.tight_layout()
	plt.show()
```

```python
plot(chunk_orders)
```

Note how clearly this shows the ...:
- ... overall dominance of the 'rising' `(1, 2, 3)` profile
- ... score positions in which that changes (development and coda)
- ... alternating pairs of triples in those divergent passages (132 and 213, then 213 and 231) which do indeed form part of longer groupings.
- ... the point at which the 132-and-213 switches to 213-and-231, which is highest point and moment where the contour of the development arpeggiation changes (bar 36).

## Another example and an Exercise

As an example of using figuration to shape the musical structure,
the Beethoven case study is perhaps especially strong,
but it is not unique: this is a much wider practice.

By way of another example, consider
_Le cygne_ ("The Swan") from
_The Carnival of the Animals_ by Camille Saint-Saëns.

The piano accompaniment is a gently lilting double figure with the typically patterns as follows:

| Hand  | Contour shape    | Notes in figure | Pitches in fig. | Per-note duration | Figure duration          |
|-------|------------------|-----------------|-----------------|-------------------|--------------------------|
| right | 1, 2, 4, 3       | 4               | 4               | 1/16th note       | 1/6 of bar (1/4 note)    |
| left  | 1, 2, 3, 2, 1, 2 | 6               | 4               | 1/8th note        | 1/2 of bar (3x 1/4 note) |

Again, changes to this pattern are highly noticeable.
