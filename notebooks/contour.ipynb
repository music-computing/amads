## <font color='crimson'> Solution 3: Explore Fitting

```python
# First a basic plot, linear only
from scipy import stats
from matplotlib import pyplot as plt
import numpy as np

def fit_and_plot(x, y):
    res = stats.linregress(x, y)
    plt.plot(x, y, 'o', label='original data')
    plt.plot(
        x,
        res.intercept + res.slope*np.array(x),
        'r', label=f'fit. $r^2$={res.rvalue.round(3)}')
    plt.xlabel("Time (proportional, semi-symbolic)")
    plt.ylabel("Pitch (MIDI number)")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
fit_and_plot(parsifal_esque_time, parsifal_esque_pitch)
```
